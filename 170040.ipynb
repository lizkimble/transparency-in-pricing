## Transform for Bounty

```
newHeaders = {"hospital_tin", "hospital_ccn", "code_meta", "description",
    "procedure_code", "code_type", "code", "modifier", "ndc", 
   "rev_code", "rev_desc", "apc", "billing_class", "patient_class", 
   "billed_quantity", "quantity_desc", "payer_desc", "payer_category",
    "payer_name", "plan_desc", "plan_name", "plan_id", 
   "is_medicare_adv", "rate", "rate_method", "rate_desc", "filename", 
   "url", "permalink"};
```

```
newData = Association[];
```

```
newData["code_meta"] = data["CodeType"];
newData["description"] = data["ProcedureDescription"];
newData["procedure_code"] = data["Procedure"];
newData["code"] = data["Code"];
newData["quantity_desc"] = data["Quantity"];
```

```
newData["code_type"] = 
  MapThread[
   Which[SCQ[ToString[#2], "HCPCS"] || SCQ[ToString[#2], "CPT"] || 
      SCQ[ToString[#1], "CPT"] || SCQ[ToString[#2], "ASA"], 
     "hcpcs_cpt", SCQ[ToString[#2], "MS-DRG"], "ms-drg", True, 
     "other"] &, Values[data[[{"CodeType", "Code"}]]]];
```

```
newData["ndc"] = (If[StringLength[#] > 10, #, "na"] & /@ data["NDC"]);
newData["rev_code"] = ((If[StringLength[#] >= 4, StringTake[#, 4], "na"]) & /@ data["RevCode"]);
newData["rev_desc"] = ((If[StringLength[#] > 7, StringDrop[#, 7], "na"]) & /@ data["RevCode"]);
newData["billing_class"] = (Which[# == "HB", "facility", # == "PB", "professional", True, "na"] & /@ data["BillType"]);
newData["billed_quantity"] = (If[StringQ[#1], ToExpression[StringReplace[StringSplit[#1, " "][[1]], "," -> ""]], #1] &) /@ data["Quantity"];
```

```
newData["rate_method"] = (StringTake[#, UpTo[500]] & /@ data["IPPricingDetail"]);
```

```
newData["filename"] = Table["48-1201402_TheUniversityOfKansasHealthSystemKansasCityCampus_standardcharges.csv", nRows];
newData["hospital_tin"] = Table["481202402", nRows];
newData["hospital_ccn"] = Table[ccn, nRows];
newData["modifier"] = Table["na", nRows]; 
newData["apc"] = Table["na", nRows];
newData["file_last_modified"] = Table["11/10/2022", nRows];
newData["hospital_ccn"] = Table[ccn, nRows];
newData["hospital_tin"] = Table["481202402", nRows];
newData["url"] = Table["https://ukhspricetransparency.blob.core.windows.net/ukhs-price-transparency/48-1201402_TheUniversityOfKansasHealthSystemKansasCityCampus_standardcharges.csv", nRows];
newData["is_medicare_adv"] = Table[0, nRows];
newData["rate_desc"] = Table["na", nRows];
newData["permalink"] = Table["null", nRows];
```
